```python
import pandas as pd

df = pd.read_csv('../data/iwildcam_2022_cameratrapai_pt_train.csv')
```

```python
gt_columns = [col.replace("gt_", "") for col in df.columns if col.startswith('gt_')]
logit_columns = [col.replace("logit_", "") for col in df.columns if col.startswith('logit_')]
print(set(gt_columns) - set(logit_columns))
print(set(logit_columns) - set(gt_columns))
```

```python
df["true_label"] = ""
for col in df.columns:
    if col.startswith("gt_"):
        df.loc[df[col], "true_label"] = col.replace("gt_", "")
```

```python
print("Len before", len(df))
df = df[df["true_label"].isin(logit_columns)]
print("Len after", len(df))
```

```python
max_logits = df[[col for col in df.columns if col.startswith("logit_")]].max(axis=1)

for col in df.columns:
    if col.startswith("logit_"):
        mask = df[col] == max_logits
        df.loc[mask, "pred_label"] = col.replace("logit_", "") + "(" + df.loc[mask, col].astype(str) + ")"
```

```python
df = df.drop(columns=[col for col in df.columns if col.startswith("gt_")])
df = df.drop(columns=[col for col in df.columns if col.startswith("logit_")])
```
